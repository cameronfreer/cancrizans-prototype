# Quick Start Guide

Get started with Cancrizans in 5 minutes!

```python
from music21 import stream, note
from cancrizans import retrograde, invert, mirror_canon

print("✓ Cancrizans imported successfully!")
```

## Create a Simple Canon

```python
# Create a theme
theme = stream.Stream()
for pitch in [60, 62, 64, 65]:
    theme.append(note.Note(pitch, quarterLength=1))

# Create a retrograde (crab) canon
crab_canon = mirror_canon(theme)

print(f"✓ Created canon with {len(crab_canon.parts)} parts")
print(f"  Duration: {crab_canon.duration.quarterLength} quarters")
```
